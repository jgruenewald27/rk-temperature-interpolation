```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the CSV file
file_path = '../RK_Results/All_Timestamps_Metrics_lrm.csv'  # Replace with your file path
df = pd.read_csv(file_path, sep=";")
df
```

```python
# Plot chour vs cv_* variables
plt.figure(figsize=(10, 6))
plt.plot(df['chour'], df['cv_rmse'], label='cv_rmse')
plt.plot(df['chour'], df['cv_mae'], label='cv_mae')
plt.plot(df['chour'], df['cv_bias'], label='cv_bias')
plt.xlabel('chour')
plt.ylabel('CV Metrics')
plt.title('chour vs CV Metrics')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot chour vs rk_* variables
plt.figure(figsize=(10, 6))
plt.plot(df['chour'], df['rk_rmse'], label='rk_rmse')
plt.plot(df['chour'], df['rk_mae'], label='rk_mae')
plt.plot(df['chour'], df['rk_bias'], label='rk_bias')
plt.xlabel('chour')
plt.ylabel('RK Metrics')
plt.title('chour vs RK Metrics')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
